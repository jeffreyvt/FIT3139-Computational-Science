```python
"""
Author: JiaHui (Jeffrey) Lu
Student ID: 25944800
"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Assumption: the guy is hooked up to a catheter, hence he always pees straight into
a bag continuously. The amount of drug in Urine is modeled in terms of the drugs in
the catheter.

The patient takes 50 milligram of drugs every 6 hours (24 15-mins long intervals.)

Note: the 
"""

k_a = 0.1
V_d = 70
delta_t = 1/4
half_life = 10
k_e = 0.693/half_life

intervals = int(192/delta_t)
```

```python
I, P, U = [0], [0], [0]
x = np.arange(1, intervals, 1)
for i in range(len(x) - 1):
    if i % 24 == 0:
        I.append(I[i] + 50 - k_a * I[i] * delta_t)
    else:
        I.append(I[i] - k_a * I[i] * delta_t)
        
    
    P.append(P[i] + k_a * I[i] * delta_t - k_e * P[i] * delta_t)
    U.append(U[i] + k_e*P[i]*delta_t)
```

```python
plt.plot(x, I, label="Intestine")
plt.plot(x, P, label="Plasma")
plt.plot(x, U, label="Urine")
plt.ylim((0,200))
plt.legend()
plt.show()
```
