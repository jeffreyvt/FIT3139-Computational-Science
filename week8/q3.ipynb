```python
"""
Author: JiaHui (Jeffrey) Lu
ID: 25944800
"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf(x):
    return np.sqrt(2/np.pi*np.exp(-x**2/2))
```

```python
fig=plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
x = np.arange(0, 10, 0.1)
y = pdf(x)
plt.plot(x, y)
plt.show()
```

```python
def expo(x, lmbda=1):
    return lmbda*np.exp(-x*lmbda)
```

```python
fig=plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
x = np.arange(0, 10, 0.1)
y = pdf(x)
c = 2.5
lmbda=1
y2 = expo(x, lmbda=lmbda)*c
plt.plot(x, y, label="half normal")
plt.plot(x, y2, label="expo")
plt.legend()
plt.show()
```

```python
c = 2.5
lmbda = 1
generated = []
n = 100000
ps = np.random.exponential(1/lmbda, n)
for i in range(len(ps)):
    yi = np.random.uniform(0, c*expo(ps[i], lmbda=lmbda))
    if yi < pdf(ps[i]): # accept
        generated.append(ps[i])
print("efficiency: ", len(generated)/n)
fig=plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
n, bins, patches = plt.hist(generated, 50, alpha=0.5, label="generated")
x = np.arange(0, 5, 0.1)
y = pdf(x)
plt.plot(x, y/y[0]*n[0], "ro", label="scaled actual distribution")
plt.legend()
plt.show()
```

##Comment:
It appears that the generated distribution has the density very similar to the actual density. Furthermore, for our given $c = 3$ and $f_P(x)=\lambda*\exp^{-\lambda x}$ where $\lambda = 0.3$ gives ~50% efficiency (50% rejected, 50% accepted)
