```python
"""
Modelling Moth

Findings:
Larger population has greater rate of change, see N_0 = 100 
compare to N_0 = 1000.
All populations converge to the carrying capacity. 
Which is between 12 and 13.
"""
lmbda, b = 1.3, 0.1
N0 = 100
x = np.arange(0, 50, 1)
N_k = [N0]
for i in range(len(x)-1):
    N_k.append(lmbda*N_k[i]/((1+N_k[i])**b))
plt.subplot(221)
plt.title('N_0 = 100')
plt.plot(x, N_k, ".")

N0 = 10
N_k = [N0]
for i in range(len(x)-1):
    N_k.append(lmbda*N_k[i]/((1+N_k[i])**b))
plt.subplot(222)
plt.title('N_0 = 10')
plt.plot(x, N_k, ".")

N0 = 1000
N_k = [N0]
for i in range(len(x)-1):
    N_k.append(lmbda*N_k[i]/((1+N_k[i])**b))
plt.subplot(223)
plt.title('N_0 = 1000')
plt.plot(x, N_k, ".")

N0 = 1
N_k = [N0]
for i in range(len(x)-1):
    N_k.append(lmbda*N_k[i]/((1+N_k[i])**b))
plt.subplot(224)
plt.plot(x, N_k, ".")
plt.title('N_0 = 1')
plt.show()
```
