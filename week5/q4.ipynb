```python
"""
Part a)

Findings: There are no difference for varying N_0
for r = 0.2, the population converges to the carrying capacity
for r = 2.3, the population oscillates around the carrying capacity
for r = 3, the model breaks down and fail to converge
"""
r = [0.5, 2.3, 3]
K = 1000
N_0s = [99, 100, 101]
x = np.arange(0, 100, 1)
for k in range(len(r)):
    for i in range(len(N_0s)):
        N_k = [N_0s[i]]
        for j in range(len(x) - 1):
            N_k.append(N_k[j] + r[k] * N_k[j] * (1 - N_k[j] / K))
        plt.subplot(3, 3, i + 1+len(r)*k)
        plt.title("N_0 = " + str(N_0s[i]) + ", r = " + str(r[k]))
        plt.plot(x, N_k)
plt.show()
```
