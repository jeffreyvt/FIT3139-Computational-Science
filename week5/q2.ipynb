```python
"""
Part a)
make right hand side of the equations equal:
    N_k * exp(a*(1-N_k/k)) = N_k + R(N_k)*N_k
    Now, solve for R(N_k).
    we get: R(N_k) = exp(a(1-N_k/K)) - 1
    
    The growth rate is initially positive, and the rate decreases as N_k increases
"""

a, K = 1, 50
N_k = np.arange(0, 50, 0.01)
R = np.exp(a*(1-N_k/K)) - 1
r = 1
P_t = np.arange(0, 50, 1)
Discrete_logistic = -(r/K)*P_t + r
plt.plot(N_k, R)
plt.plot(P_t, Discrete_logistic, ".")
plt.show()
"""
Part b
As you can see, the discrete logistic growth rate decays in a linear fashion, where as the continuous model
decreases exponentially as the carrying capacity is reached.

Furthermore, due to the fundamental difference between discrete, and continuous, the continuous model's rate is 
smooth, and its instantaneous rate can be calculated. On the other hand, the discrete model is... discrete.

Part c
The unrestricted growth rate occurs at N_k = 0
R(0) = exp(a) - 1 = r
solving the equation for a, we get: a = ln(r+1)
"""
```

```python
x = np.arange(0, 10, 1)
y = [100]
K = 1000
a = 1.8
for i in range(len(x)-1):
    y.append(y[i]*np.exp(a*(1-y[i]/K)))
plt.plot(x, y)
plt.show()

"""
d) i)
The oscillation with periodicity of 2 appears for 1<a<2
"""
```
