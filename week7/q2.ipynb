```python
def dIdt(s, i, f=0.001, a=0.01):
    return f*s*i-a*i


def dSdt(i,s, f=0.001):
    return -f*s*i
```

```python
def euler(f, h, x_0, y_0):
    return y_0 + h * f(x_0, y_0)
```

```python
"""
The following code plots the graph for the Susceptible vs Infective simulation.

"""

h = 0.1
t_s = np.arange(0, 100, h)
S_vals = [1000]
I_vals = [5]
for i in range(len(t_s)-1):
    S_vals.append(euler(dSdt, h, I_vals[i], S_vals[i]))
    I_vals.append(euler(dIdt, h, S_vals[i], I_vals[i]))


fig=plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.ylim((0, 1000))
plt.plot(t_s, S_vals, label="Susceptibles")
plt.plot(t_s, I_vals, label="Infectives")
plt.legend()
plt.show()
```
